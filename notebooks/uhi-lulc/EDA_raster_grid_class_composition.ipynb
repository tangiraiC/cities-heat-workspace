```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
import altair as alt
alt.data_transformers.disable_max_rows()
alt.__version__
```

```python
df = pd.read_csv('class_area_example1.csv') #class_area_example2, class_area_example3

map_dict = {"fid":"fid","0":"open_space",
  "1":"nonresidential",
  "2":"atomistic",
  "3":"informal_subdivision",
  "4":"formal_subdivision",
  "5":"housing_project",
  "6":"road"}

df = df.rename(columns=map_dict)
row_count = len(df)
g = df.melt()
g = g.loc[~(g.variable =='fid')]
```

The Frequency histograms look at the distribution of ULU class areas contained in the grid cells of the 300m UHI data for a single city, with area in hectares along the x axis of each facet, and frequency count in the y axis. If we consider a 300m^2 cell 9 hectares, completely homogenous class neighborhoods show up towards the 8-9

```python
alt.Chart(g, width=100,height=80).mark_bar().encode(
    alt.X("value:Q", bin=True, title="hectares"),
    y=alt.Y('count()',scale=alt.Scale(domain=[0, row_count]), title="frequency in sample"),
).facet(
    facet=alt.Facet('variable:N', title=None), title="Class area overlap"
)
```

```python
df = df.set_index('fid')
```

```python
# How many 
thresh = 8
df[df.gt(thresh).any(axis=1)]
```

```python
df_minus_mode = df.mask(
    df.eq(df.max(axis=1), axis=0), 
    np.nan,
    axis=0)
```

```python
print("Total area (ha) per class overlapping with UHI grid cells:")
print(df.sum())
print("\n---\n")
print("Total area (ha) per class dropped if reducing to modal class:")
print(df_minus_mode.sum())
print("\n---\n")
print("Fraction of class area dropped if reducing to modal class:")
print(df_minus_mode.sum()/df.sum())
```
